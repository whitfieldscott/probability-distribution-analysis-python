## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Normal Distributions

```python
from scipy.stats import norm

def dnorm(x, mean=0, sd=1):
    """
    Calculates the point density (PDF) of the normal distribution
    """
    result = norm.pdf(x, loc=mean, scale=sd)

    return result

def pnorm(q, mean=0, sd=1):
    """
    Calculates the cumulative distribution (CDF) of the normal distribution
    """
    result = norm.cdf(q, loc=mean, scale=sd)

    return result

def qnorm(p, mean=0, sd=1):
    """
    Calculates the quantile function (inverse CDF) from the normal distribution
    """
    result = norm.ppf(p, loc=mean, scale=sd)

    return result

def rnorm(n, mean=0, sd=1):
    """
    Generates random variables from the normal distribution
    """
    result = norm.rvs(loc=mean, scale=sd, size=n)

    return result
```

```python
print(f"The Point Density of a Normal Distribution is (0, 1) at x = 2: {dnorm(2, mean=0, sd=1)}")
print(f"The Cumulative Distribution is (0, 1) at x = 2: {pnorm(2, mean=0, sd=1)}")
print(f"Which value of a Normal(0, 1) has a cumulative probability of 0.9? {qnorm(0.9, mean=0, sd=1)}")
print(f"Generate 2000 random values from a Normal(0, 1): {rnorm(2000, mean=0, sd=1)}")
```

## Normal Distribution Plot Chart

```python
from scipy.stats import norm

mu, sigma = 0, 1

data = np.linspace(-5, 5, 2000)
pdf = norm.pdf(data, mu, sigma)

plt.figure(figsize = (10, 5))
plt.plot(data, pdf)
plt.title("Uniform Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

## Uniform Distributions

```python
from scipy.stats import uniform

def duniform(x, min_val=0, max_val=1):
    """
    Calculates the point density (PDF) of the uniform distribution
    """
    result = uniform.pdf(x, loc=min_val, scale=max_val - min_val)

    return result

def puniform(q, min_val=0, max_val=1):
    """
    Calculates the cumulative distribution (CDF) of the uniform distribution
    """
    result = uniform.cdf(q, loc=min_val, scale=max_val - min_val)

    return result

def quniform(p, min_val=0, max_val=1):
    """
    Calculates the quantile function (inverse CDF) from the uniform distribution
    """
    result = uniform.ppf(p, loc=min_val, scale=max_val - min_val)

    return result

def runiform(n, min_val=0, max_val=1):
    """
    Generates random variables from the uniform distribution
    """
    result = uniform.rvs(loc=min_val, scale=max_val - min_val, size=n)

    return result
```

```python
print(f"The Point Density of a Uniform Distribution is (0, 1) at x = 2: {duniform(2, min_val=0, max_val=1)}")
print(f"The Cumulative Distribution is (0, 1) at x = 2: {puniform(2, min_val=0, max_val=1)}")
print(f"Which value of a Uniform(0, 1) has a cumulative probability of 0.9? {quniform(0.9, min_val=0, max_val=1)}")
print(f"Generate 2000 random values from a Uniform(0, 1): {runiform(2000, min_val=0, max_val=1)}")
```
